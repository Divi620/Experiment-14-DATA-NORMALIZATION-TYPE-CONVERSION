```python
#Dataset
import pandas as pd
import numpy as np

data={
    'Product':['laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],
    'Price': [55000,20000,30000,60000,45000,50000], # Added 50000 for Camera
    'Units Sold': [100,50,75,200,500,320],
    'Discount': [10,15,12,20,18,8]}

df = pd.DataFrame(data)
print('Original dataset')
print(df)
```

```python
#1. Min-Max
df['Price_MinMax'] = (df['Price'] - df['Price'].min()) / (df['Price'].max() - df['Price'].min())
print('Min-Max Normalized dataset')
print(df[['Product','Price','Price_MinMax']])
```

```python
#2. Z-score Normalization
df['Price_ZScore'] = (df['Price'] - df['Price'].mean()) / df['Price'].std()
print('Z-score Normalized dataset')
print(df[['Product','Price','Price_ZScore']])
```

```python
#3. Decimal Scaling
df['Price_DecimalScaling'] = df['Price'] /100000
print('Decimal Scaled dataset')
print(df[['Product','Price','Price_DecimalScaling']])
```

```python
df['Discount_DecimalScaling'] = df['Discount'] / 100
print('Decimal Scaled Discount dataset')
print(df[['Product','Discount','Discount_DecimalScaling']])
```

```python
#Normalise Multiple Columns
cols=['Price','Units Sold','Discount']
df_norm=(df[cols]-df[cols].min())/(df[cols].max()-df[cols].min())
print('\nNormalized dataset')
print(df_norm)
```

Part-B TURNING CATEGORICAL VARIABLES INTO QUANTITATIVE VARIABLES IN PYTHON

```python
import pandas as pd
import numpy as np
data = {
    'Order_ID': [101,102,103,104,105,106],
    'Customer_Gender':['Male','Female','Female','Male','Female','Male'],
    'Product_Category':['Electronics','Clothing','Electronics','Clothing','Electronics','Home'],
    'Payment_Method':['Cash','Credit Card','Cash','Credit Card','Cash','Credit Card'],
    'City':['Pune','Mumbai','Pune','Mumbai','Pune','Mumbai'],
    'Order_Value':[15000,2300,3200,5200,1800,4100]}

df = pd.DataFrame(data)
print(df)
```

```python
#Method 1: Label Encoding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Gender_Label'] = label_encoder.fit_transform(df['Customer_Gender'])
print(df)
```

```python
#Method 2: One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=['Payment_Method'])
print(df_encoded)
```

```python
#Method 3: Encoding Multiple columns
df_multi = pd.get_dummies(df, columns=['Customer_Gender','Product_Category'])
print(df_multi)
```

```python
#Method 4: dummy encoding (Drop First)
df_dummy = pd.get_dummies(df, columns=['Customer_Gender'], drop_first=True)
print(df_dummy)
```
